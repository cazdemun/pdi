```python
import cv2
# Espacio de color HSV

# H -> angulos (0, 179)
# S -> saturacion (0, 255)
# V -> intensidad (0, 255)

img = np.zeros((360, 256, 3), np.uint8)

for interval in range(180):
    min_int = interval * 2
    max_int = (interval + 1) * 2
    img[min_int:max_int, :, :] = [interval, 255, 255]
    
out_img = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(20, 10))
plt.imshow(out_img)
plt.show()
```
