```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# Deteccion de celulas cancerigenas
img = cv2.imread('P6070001.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(20, 10))
plt.imshow(img_rgb)
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_converted = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)


img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb2 = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)


plt.figure(figsize=(20, 10))
plt.imshow(img_rgb2)
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_blue = np.array([0, 0, 230])
upper_blue = np.array([10, 255, 255])

mask = cv2.inRange(img_hsv, lower_blue, upper_blue)
mask = cv2.GaussianBlur(mask, (3,3), 0)

res = cv2.bitwise_and(img_rgb, img_rgb, mask = mask)

plt.figure(figsize=(20, 10))
img_converted = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)
#plt.imshow(img_converted)
plt.imshow(res)
plt.show()
```

```python
print(res.shape)
```

```python
print(res.shape)

def distance(u, v, img):
    M, N, _ = img.shape
    d = np.sqrt(np.power(u - M / 2, 2) + np.power(v - N / 2, 2))
    return d

def butter(u, v, d0, n, img):
    dummy = 1 / (1 + np.power(distance(u, v, img)/d0, 2*n))
    return dummy

def customMask(img):
    rows, cols, _ = img.shape
    mask = np.ones((rows, cols))
    for x in range(0,rows):
        for y in range(0,cols):
            mask[x][y] = butter(x, y, 50, 6, img)
    return mask
    
mask = customMask(res)
# Applying filters
img_gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
img_gray = img_gray * mask
#print(img_gray.shape)
#img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
#print(img_gray.shape)
#img_gray = cv2.cvtColor(img_gray, cv2.COLOR_RGB2GRAY)
#print(img_gray)
# Printing
plt.figure(figsize=(20, 10))
plt.imshow(img_gray, 'gray')
plt.show()
```

```python
row, col, _ = img.shape
pad = col - row 
x = cv2.getGaussianKernel(row, 100)
print(x.shape)
gaussian = x * x.T

print(gaussian)

superGaussian = gaussian

for i in range(0, row):
    for j in range(0, row):
        if gaussian[i][j] < 0.000005:
            superGaussian[i][j] = 0
        else:
            superGaussian[i][j] = 1

mask = np.zeros((row, col))

mask[:row,pad//2:pad//2+row] = superGaussian

# Applying filters
img_gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
img_gray = img_gray * mask
#print(img_gray.shape)
#img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
#print(img_gray.shape)
#img_gray = cv2.cvtColor(img_gray, cv2.COLOR_RGB2GRAY)
#print(img_gray)
# Printing
plt.figure(figsize=(20, 10))
plt.imshow(img_gray, 'gray')
plt.show()
```
